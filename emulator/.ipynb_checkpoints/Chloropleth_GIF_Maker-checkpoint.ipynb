Create a GIF of a particular fire

```python
import pandas as pd
import os
import rasterio
import matplotlib.pyplot as plt
import numpy as np
from copy import copy

pick_dir = cwd = os.path.dirname(os.path.dirname(os.getcwd())) + '/data/'

data_pkl = pick_dir + 'master.pkl'

dataframe = pd.read_pickle(data_pkl)
idx = 9  # Sample number

arrival_f = dataframe.loc[idx].arrival_time_proc



with rasterio.open(arrival_f) as src:
    img = src.read()
    img = img[-1,:,:]

    # invert image
    img = np.where(img > 1, 24-img, -1.0)

    plt.figure(figsize=(12,12))
    plt.imshow(img, cmap='plasma')
    plt.colorbar()
```

```python
interval = 12

import cv2
import os
from matplotlib import cm
from PIL import Image
from array2gif import write_gif

from matplotlib.animation import FuncAnimation, PillowWriter

width, height = img.shape

cmap = 'plasma'

# create colormap palette
palette = copy(plt.get_cmap(cmap))
#palette.set_under('white', '1.0') # set opaque white for unburnt
palette.set_under(alpha = 0.0)



# gets an image at the appropriate time interval
def make_image(img, interval):

    img_interval = np.where(img <0, 23, img)
    img_interval = img_interval - interval

    rgba_arr = cm.palette(img_interval, bytes=True) # hot
    # add transparency
    alpha = np.where(img_interval > 1.001, 0, 127)
    rgba_arr[:,:,-1] = alpha

    # remove alpha channel
    #rgb_arr = rgba_arr[:,:,:-1]
    rgb_arr = rgba_arr
    #rgb_arr = np.moveaxis(rgb_arr, -1, 0)
    
    
    
    rgb_img = Image.fromarray(rgb_arr)
    rgb_img = rgb_img.convert("RGBA")
    
    # convert to image
    #rgba_img = Image.fromarray(rgba_arr)
    #rgba_img = rgba_img.convert("RGBA")
    return rgb_img
    


time_slices = np.arange(0, 21, 0.1) # 3 minute increments

gif_img = Image.new('RGBA', (width, height))

# array of arrays
frames = []

for time_slice in time_slices:
    
    frame = make_image(img, time_slice)
    frames.append(frame)
    
frames[0].save('chloropleth.gif', format='GIF',
               append_images=frames[1:],
               save_all=True,
               duration=30, loop=0, transparency=0)
```
