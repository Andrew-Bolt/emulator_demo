## 4. Create tensorflow Dataset objects

These objects are used by tensorflow to enable efficient batching, caching and distribution of trials during training or predicting

```python
BUFFER_SIZE = 20
REPEAT = 1 # 37
BATCH_SIZE = 16# 64 # For batch > 1 all tensors must have the same shape (no transposes)
load_trial = ft.dataset.load_trial
load_arrival = ft.dataset.load_arrival
load_predict = ft.dataset.load_predict
load_start_stop_predict = ft.dataset.load_start_stop_predict
load_lc = ft.dataset.load_lc
load_half = ft.dataset.load_half
transform_trial = ft.dataset.transform_trial
transform_test = ft.dataset.transform_test
transform_identity = ft.dataset.transform_identity

START_INTERVAL = 0
STOP_INTERVAL = 22

window = 512#256

random.seed(42)


def train_generator(window=window, crop=True):
    
    # randomly order the trails to process each epoch
    train_indices = trials_df[trials_df.train_or_test == 'train'].index
    indices = np.random.choice(train_indices, len(train_indices), replace=False)
    
    for ind in indices:
        
        features, target = load_trial(ind, trials_df, window_size=window, crop=crop, transform=[0.5, 0.5, 0.5, 0.0]) 
        
        yield features, target

def predict_generator(window=None, crop=False):
    
    # randomly order the trials to process each epoch
    predict_indices = master_df.index # search over all possible entries

    indices = np.random.choice(predict_indices, len(predict_indices), replace=False)
    
    for ind in indices:
        
        features, target = load_predict(ind, master_df, transform = [0.0, 0.0, 0.0, 0.0])
        
        yield features, target
        
def predict_start_stop_generator(window=None, crop=False, start=START_INTERVAL, stop=STOP_INTERVAL):
    
    # randomly order the trials to process each epoch
    predict_indices = master_df.index # search over all validation entries
    #predict_indices = trials_df[trials_df.train_or_test == 'test'].index

    #indices = predict_indices
    indices = np.random.choice(predict_indices, len(predict_indices), replace=False)
    
    for ind in indices:
        
        features, target = load_start_stop_predict(ind, master_df, transform = [0.0, 0.0, 0.0, 0.0], start=START_INTERVAL, stop=STOP_INTERVAL)
        
        yield features, target
        
        
        
def half_generator(window=None, crop=False):
    
    # randomly order the trials to process each epoch
    predict_indices = master_df.index # search over all possible entries

    indices = np.random.choice(predict_indices, len(predict_indices), replace=False)
    
    for ind in indices:
        
        features, target = load_half(ind, master_df, transform = [0.0, 0.0, 0.0, 0.0])
        
        yield features, target

def autoenc_generator(window=window, crop=True):
    
    # randomly order the trails to process each epoch
    trial_indices = trials_df[trials_df.train_or_test == 'train'].index
    
    indices = np.random.choice(trial_indices, len(trial_indices), replace=False)
    
    
    for ind in indices:
        
        features, target = load_arrival(ind, trials_df, window_size=window, crop=crop) 
        
        yield features, target
  
def autoenc_topo_generator(window=window, crop=True):
    
    # randomly order the trails to process each epoch
    trial_indices = trials_df[trials_df.train_or_test == 'train'].index
    
    indices = np.random.choice(trial_indices, len(trial_indices), replace=False)
    
    
    for ind in indices:
        
        features, target = load_lc(ind, trials_df, window_size=window, crop=crop) 
        
        yield features, target


def test_generator(window=window, crop=True):
    
    # randomly order the trails to process each epoch
    test_indices = trials_df[trials_df.train_or_test == 'test'].index
    
    indices = np.random.choice(test_indices, len(test_indices), replace=False)
    
    
    for ind in indices:
        
        features, target = load_trial(ind, trials_df, window_size=window, crop=crop, transform=[0.5, 0.5, 0.5, 0.0]) 
        
        yield features, target
 

# TRAIN DATASETS
# train dataset is based on 'trials' dataframe
# training occurs on single interval (fine is RNN is memoryless)

output_types   = ((tf.float32, tf.float32, tf.float32, tf.float32, tf.float32, tf.float32), tf.float32)
ae_output_types = (tf.float32, tf.float32)
ae_topo_output_types = (tf.float32, tf.float32)
output_shapes  = (
(tf.TensorShape([None, None, None]), 
 tf.TensorShape([None, 8]), 
 tf.TensorShape([2]),
 tf.TensorShape([None, None, None]), 
 tf.TensorShape([None, None, None]),
 tf.TensorShape([4])),
 tf.TensorShape([None, None, None]),
)
ae_output_shapes = (tf.TensorShape([None, None, None]), tf.TensorShape([None, None, None]))
ae_topo_output_shapes = (tf.TensorShape([None, None, None]), tf.TensorShape([None, None, None]))
padded_shapes  = (((None,None,None), (None, 8), (2,), (None,None,None), (None,None,None), (4,)),(None,None,None))

train_dataset = tf.data.Dataset.from_generator(train_generator, output_types=output_types, output_shapes = output_shapes) 
train_dataset = train_dataset.repeat(REPEAT)
train_dataset = train_dataset.map(transform_trial, num_parallel_calls=tf.data.experimental.AUTOTUNE) 
train_dataset = train_dataset.shuffle(BUFFER_SIZE)
train_dataset = train_dataset.batch(BATCH_SIZE, drop_remainder=True)
#train_dataset = train_dataset.padded_batch(BATCH_SIZE, padded_shapes=padded_shapes)

test_dataset = tf.data.Dataset.from_generator(test_generator, output_types=output_types, output_shapes = output_shapes) 
test_dataset = test_dataset.repeat(REPEAT)
test_dataset = test_dataset.map(transform_trial, num_parallel_calls=tf.data.experimental.AUTOTUNE) 
test_dataset = test_dataset.shuffle(BUFFER_SIZE)
test_dataset = test_dataset.batch(BATCH_SIZE, drop_remainder=True)
#test_dataset = test_dataset.padded_batch(BATCH_SIZE, padded_shapes=padded_shapes)

  
# PREDICT DATASETS
# predict dataset is based on 'master' dataframe
# predictions occur across all intervals

predict_dataset = tf.data.Dataset.from_generator(predict_generator, output_types=output_types, output_shapes = output_shapes)
predict_dataset = predict_dataset.map(transform_trial, num_parallel_calls=tf.data.experimental.AUTOTUNE)
predict_dataset = predict_dataset.batch(1)

predict_start_stop_dataset = tf.data.Dataset.from_generator(predict_start_stop_generator, output_types=output_types, output_shapes = output_shapes)
predict_start_stop_dataset = predict_start_stop_dataset.map(transform_trial, num_parallel_calls=tf.data.experimental.AUTOTUNE)
predict_start_stop_dataset = predict_start_stop_dataset.batch(1)

half_dataset = tf.data.Dataset.from_generator(half_generator, output_types=output_types, output_shapes=output_shapes)
half_dataset = half_dataset.map(transform_trial, num_parallel_calls=tf.data.experimental.AUTOTUNE)
half_dataset = half_dataset.batch(1)

# AUTO ENCODE DATASET

autoenc_dataset = tf.data.Dataset.from_generator(autoenc_generator, output_types = ae_output_types, output_shapes = ae_output_shapes)
autoenc_dataset = autoenc_dataset.map(transform_identity, num_parallel_calls=tf.data.experimental.AUTOTUNE)
autoenc_dataset = autoenc_dataset.batch(BATCH_SIZE, drop_remainder=True)

# AUTO ENCODE TOPOGRAPHY DATASET
    # included so these weights do not change during main training
    # hopefully removes the possibility that components in this path are used to update fire position
    # also makes emulator training quicker/more precise since fewer weights need to be learned
    
autoenc_topo_dataset = tf.data.Dataset.from_generator(autoenc_topo_generator, output_types = ae_topo_output_types, output_shapes = ae_topo_output_shapes)
autoenc_topo_dataset = autoenc_topo_dataset.map(transform_identity, num_parallel_calls=tf.data.experimental.AUTOTUNE)
autoenc_topo_dataset = autoenc_topo_dataset.batch(BATCH_SIZE, drop_remainder=True)
```

```python
# confirm the dataset is build correctly (input elements with correct shape)

if True:
    for element in train_dataset:
        print(element[0][0].shape) # arrival init
        print(element[0][1].shape) # weather
        print(element[0][2].shape) # climate
        print(element[0][3].shape) # heightmap
        print(element[0][4].shape) # landclass
        
        print(element[1].shape)    # target
        break
print('----')

if False:
    for element in test_dataset:
        print(element[0][0].shape) # arrival init
        print(element[0][1].shape) # weather
        print(element[0][2].shape) # heightmap
        print(element[0][3].shape) # landclass
        
        print(element[1].shape)    # target
        break
print('----')

if False:
    for element in predict_dataset:
        print(element[0][0].shape) # arrival init
        print(element[0][1].shape) # weather
        print(element[0][2].shape) # heightmap
        print(element[0][3].shape) # landclass
        
        print(element[1].shape)    # target
        break
print('----')

if False:
    for element in predict_start_stop_dataset:
        print(element[0][0].shape) # arrival init
        print(element[0][1].shape) # weather
        print(element[0][2].shape) # heightmap
        print(element[0][3].shape) # landclass
        
        print(element[1].shape)    # target
        break
print('----')

if False:
    for element in autoenc_dataset:
        print(element[0].shape) # arrival

        print(element[1].shape)    # target
        break
print('----')

if False:
    for element in autoenc_topo_dataset:
        print(element[0].shape) # arrival

        print(element[1].shape)    # target
        break
print('----')

if False:
    for element in half_dataset:
        print(element[0][0].shape) # arrival
        print(element[0][1].shape) # weather
        print(element[0][2].shape) # heightmap
        print(element[0][3].shape) # landclass
        print(element[1].shape)    # target
        break
print('----')
```

```python
# if Train is true the model weights are updated
# if RESET is true weights are reset, and training proceeds from scratch

TRAIN = False
RESET = False
period  = 10 # save every N epochs
EPOCHS  = 50

# set up training checkpoints
checkpoint_path = cwd + '/model_checkpoints/cp-{epoch:04d}.ckpt'
checkpoint_dir = os.path.dirname(checkpoint_path)

# number of steps per epoch
steps_per_epoch = int(np.ceil(len(trials_df[trials_df.train_or_test == 'train']) / 4))

# create a callback that saves the model's weights every 10 epochs

cp_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_path, 
    verbose=1, 
    save_weights_only=True,
    #save_freq=period*steps_per_epoch) # save every 10 epochs
    period = period)


# reset model
hsig = 'hard_sigmoid'
sig = 'sigmoid'
tanh = 'tanh'
leaky_10 = tf.keras.layers.LeakyReLU(alpha=.1)
activation = leaky_10

if RESET:
    # reset the model (comment this out to continue training from previous value)
    activation = tf.keras.layers.LeakyReLU(alpha=.1)
    
    model = ft.model.cnn_rnn_model(activation)
    # set the optimizer
    opt = tf.keras.optimizers.legacy.Adam(learning_rate=0.001) #lr = 0.001, default
    # compile the model
    model.compile(optimizer=opt, loss=loss_func, metrics=[iou_metric])
else:
    #model = tf.keras.models.load_model('./latest_model',
    #                                 custom_objects={'l1_mod_relative_loss_cropped': loss_func, 'iou_metric': iou_metric})
    
    model = ft.model.cnn_rnn_model(activation)
    opt = tf.keras.optimizers.legacy.Adam(learning_rate=0.001)
    
    model.load_weights('./latest_model/saved_weights.ckpt')
    model.compile(optimizer=opt, loss=loss_func, metrics=[iou_metric])
    

## Tensorboard setup
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)

    

# save initial model weights
##model.save_weights(checkpoint_path.format(epoch=0))

if TRAIN: 
    
    # calculate model weights
    #history = model.fit(train_dataset, validation_data=test_dataset, epochs=EPOCHS, callbacks=[cp_callback], verbose=1, use_multiprocessing=True)
    history = model.fit(train_dataset, validation_data=test_dataset, epochs=EPOCHS,
                        callbacks=[cp_callback, tensorboard_callback], verbose=1, use_multiprocessing=True)

        
    plt.figure(figsize=(12,6))
    plt.plot(history.history['loss'])
    #plt.plot(history.history['val_loss'])
    plt.title('loss')
    plt.xlabel('epoch')
    plt.ylabel('loss')
    plt.show()
    fn = Path('~/Pictures/training_loss.pdf').expanduser()
    plt.savefig(fn, bbox_inches='tight')
    
    plt.figure(figsize=(12,6))
    plt.plot(history.history['<lambda>'])
    #plt.plot(history.history['val_loss'])
    plt.title('Jaccard Score')
    plt.xlabel('epoch')
    plt.ylabel('Jaccard Score')
    plt.ylim([0,1])
    plt.show()
    fn = Path('~/Pictures/jaccard_score.pdf').expanduser()
    plt.savefig(fn, bbox_inches='tight')
    if False:
        plt.figure(figsize=(12,6))
        plt.plot(history.history['dice_metric_cropped'])
        #plt.plot(history.history['val_loss'])
        plt.title('Dice Score')
        plt.xlabel('epoch')
        plt.ylabel('Dice Score')
        plt.ylim([0,1])
        plt.show()
        fn = Path('~/Pictures/dice_score.pdf').expanduser()
        plt.savefig(fn, bbox_inches='tight')
        
    model.save_weights('./latest_model/saved_weights.ckpt')
    
else:
    
    # load model weights
    latest_ckpt = tf.train.latest_checkpoint(checkpoint_dir)
    model.load_weights('./latest_model/saved_weights.ckpt')
```

```python
def predict_start_stop_generator(window=None, crop=False, start=START_INTERVAL, stop=STOP_INTERVAL):
    
    # randomly order the trials to process each epoch
    predict_indices = master_df.index # search over all validation entries
    #predict_indices = trials_df[trials_df.train_or_test == 'test'].index

    #indices = predict_indices
    indices = np.random.choice(predict_indices, len(predict_indices), replace=False)
    
    for ind in indices:
        
        features, target = load_start_stop_predict(ind, master_df, transform = [0.0, 0.0, 0.0, 0.0], start=START_INTERVAL, stop=STOP_INTERVAL)
        
        yield features, target

folder = 'architecture results/'        
#folder = 'architecture results/trial_256c_3d_1d_64p_avgpool/'


if False: # determine whether to iterate over the predict dataset and determine evaluation metrics

    iou = ft.metric_and_loss.iou_metric
    dice = ft.metric_and_loss.dice_metric

    total_iou_vals = []
    total_dice_vals = []
    total_weighted_iou_vals = []
    total_weighted_dice_vals = []

    scope = range(0, 21)
    scope = [0, 5]
    
    #data = predict_start_stop_dataset.as_numpy_iterator()
    for j in scope:
        START_INTERVAL = j
        STOP_INTERVAL = 22

        iou_vals = []
        dice_vals = []

        weight_vals = []

        predict_start_stop_dataset = tf.data.Dataset.from_generator(predict_start_stop_generator, output_types=output_types, output_shapes = output_shapes)
        predict_start_stop_dataset = predict_start_stop_dataset.map(transform_trial, num_parallel_calls=tf.data.experimental.AUTOTUNE)
        predict_start_stop_dataset = predict_start_stop_dataset.batch(1)

        for trial in predict_start_stop_dataset.as_numpy_iterator():
            X, y = trial

            y_p = model.predict((X))

            max_val = np.max(y)
            weighting = np.sum(np.where((y > 0)&(y<max_val), 1, 0)) # burnt area
            jaccard_score = iou(y, y_p).numpy()
            dice_score = (2 * jaccard_score) / (jaccard_score + 1)

            iou_vals.append(jaccard_score)
            dice_vals.append(dice_score)
            weight_vals.append(weighting)

        #print('IOU mean : ' + str(np.average(iou_vals, weights=weight_vals)))
        #print('IOU standard deviation: ' + str(np.std(iou_vals)))

        #print('Dice mean : ' + str(np.average(dice_vals, weights=weight_vals)))
        #print('Dice standard deviation: ' + str(np.std(dice_vals)))

        total_iou_vals.append(np.mean(iou_vals))
        total_weighted_iou_vals.append(np.average(iou_vals, weights=weight_vals))

        total_dice_vals.append(np.mean(dice_vals))
        total_weighted_dice_vals.append(np.average(dice_vals, weights=weight_vals))

    #plot and save Jaccard/IOU scores
    plt.figure(figsize=(12,6))
    plt.plot(total_iou_vals)
    plt.savefig(folder + 'iou_scores.pdf', bbox_inches='tight')
    plt.show()
    
    plt.figure(figsize=(12,6))                           
    plt.plot(total_weighted_iou_vals)
    plt.savefig(folder + 'iou_scores_weighted.pdf', bbox_inches='tight')
    plt.show()
    
    #plot and save Dice scores
    plt.figure(figsize=(12,6))
    plt.plot(total_dice_vals)
    plt.savefig(folder + 'dice_scores.pdf', bbox_inches='tight')
    plt.show()
    
    plt.figure(figsize=(12,6))                           
    plt.plot(total_weighted_dice_vals)
    plt.savefig(folder + 'dice_scores_weighted.pdf', bbox_inches='tight')
    plt.show()
    
```
