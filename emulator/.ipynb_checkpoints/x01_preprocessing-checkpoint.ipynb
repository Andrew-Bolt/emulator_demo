## Preprocessing

This firetools.preprocessing module can be used to handle raw data from spark or similar platforms and engineer them into machine learning ready formats.

fire_tools.preprocessing.preprocess_data() takes raw data from a folder and saves processed data to a new folder. Filenames are stored within a .csv file, which can be called to retrieve the files associated with individual samples. 

For example a sample raw arrival time chloropleth, heightmap, landclass map and weather data may be processed and saved to a new location. A new row is appended to the master .csv file which records the location of the raw data, processed data, and a selection of statistics for the sample.


___
*csv columns*

index, sample_number,
arrival_time_raw, heightmap_raw, landclass_raw, weather_raw, 
arrival_time_proc, heightmap_proc, landclass_proc, weather_proc,
sample_class, num_intervals, date, duration, location, size(sqkm), max_wind(km/h)


sample_number: unique number generated from the raw files <br>
arrival_time_raw, ..., weather proc: the location of files associated with a sample <br>
sample class: user defined broad classiciation of the type of fire that burned <br>
num_intervals: the number of time steps used <br>
date: YYYY/mm/DD <br>
duation: HH/MM/SS <br>
location: lat/lon <br>
size: burned area is square km <br>



max_wind: maximum recorded windspeed in pixels/interval <br>
min_wind: minimum recorded windspeed in pixels/interval <br>
max_temp: maximum recorded temperature in degrees C <br>
min_temp: minimum recorded temperature in degrees C <br>
max_height: maximum height measured in pixel values <br>
min_height: minimum height measured in pixel values <br>

---

*landclass notes*

0: unburnable/water
1: FHSs
2: FHSns
3: Hns
4: mallee cover
5: mallee fuel heights
6: heathland fuel heights

### A note on GeoTIFF format

I believe that the input heightmap and landclass images are already compressed as recieved. After a CRS coordinate change, images have the same resolution and number of pixels, as well as the same data type (Uint16, Uint8 respectively) however the file size after transformation is approximately 30 times larger. This is likey due to lossless compression. For this reason it may be more useful to continue cropping out patches of these images in uncompressed (fast to read, large in space) format. 

It may also be interesting to apply compression algorithms to these patched images, since a 30x factor saving will reduce the approx 10 Gb dataset to approx 300mb. This comes at a tradeoff of a slightly higher time to read in the images - but hopefully this isn't massively important.

Uncompressed each heigtmap image is approx 25mb (1526px x 2048px), each landclass is 18.9 mb and each arrival image is 47mb. Potentially I can replace "arrival_start" and simply use an additional channel in "arrival", thus saving approx 30% space w/o altering anything of significance.
